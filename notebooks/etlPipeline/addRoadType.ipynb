```python
import pandas as pd
from doSpatialJoin import joinCSVs

joinedDf = joinCSVs("../../data/processed/dataset_v5.parquet", "../../data/raw/texas_edges.parquet", True, ["lon", "lat", "highway"], True, 100, True, False)

joinedDf["HighwayRoad"] = (
    (joinedDf["highway"] == "motorway") | (joinedDf["highway"] == "motorway_link") |
    (joinedDf["highway"] == "trunk") | (joinedDf["highway"] == "trunk_link")
).astype(int)

joinedDf["MajorUrbanRoad"] = (
    (joinedDf["highway"] == "primary") | (joinedDf["highway"] == "primary_link")
).astype(int)

joinedDf["MinorRoad"] = (
    (joinedDf["highway"] == "secondary") | (joinedDf["highway"] == "secondary_link") |
    (joinedDf["highway"] == "tertiary") | (joinedDf["highway"] == "tertiary_link") | (joinedDf["highway"] == "residential")
).astype(int)

# get all the ones that have values, invert them to label the ones that don't
joinedDf["UnknownRoad"] = (
    ~((joinedDf["HighwayRoad"] | joinedDf["MajorUrbanRoad"] | joinedDf["MinorRoad"]).astype(bool))
).astype(int)

print(f"NUMBER OF UNKNOWN ROADS: {(joinedDf["UnknownRoad"] == 1).sum()}")

# drop them if they're unknown
# joinedDf = joinedDf[joinedDf["UnknownRoad"] == 0]

joinedDf.to_parquet("../../data/processed/dataset_v6.parquet")
```
