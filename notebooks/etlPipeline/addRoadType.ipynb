```python
from doSpatialJoin import joinCSVs

joinedDf = joinCSVs("../../data/processed/dataset_v3.csv", "../../data/txdotCSVs/TxDOT_Functional_Classification.shp", ["geometry", "FC_DESC"])

joinedDf["CollectorRoad"] = (
    (joinedDf["FC_DESC"] == "Major Collector") | (joinedDf["FC_DESC"] == "Minor Collector")
).astype(int)

joinedDf["ArterialRoad"] = (
    (joinedDf["FC_DESC"] == "Principal Arterial - Other") | (joinedDf["FC_DESC"] == "Minor Arterial")
).astype(int)

joinedDf["HighwayRoad"] = (
    (joinedDf["FC_DESC"] == "Interstate") | (joinedDf["FC_DESC"] == "Principal Arterial - (Other Freeways and Expressways)")
).astype(int)

joinedDf.drop(columns=["FC_DESC"], inplace=True)

joinedDf.to_csv("../../data/processed/dataset_v4.csv")

```
